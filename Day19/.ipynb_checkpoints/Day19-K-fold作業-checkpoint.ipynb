## 作業目標：運用scikit-learn API 實現K-fold分割資料

---

### 讀入資料

```python
import pandas as pd
dataset = pd.read_csv(r'Social_Network_Ads.csv')
```

```python
dataset
```

### 取出訓練特徵與標註

```python
X = dataset[['User ID', 'Gender', 'Age', 'EstimatedSalary']].values
Y = dataset['Purchased'].values
```

---

```python
import numpy as np
from sklearn.model_selection import KFold
```

### 將訓練資料按照順序切割成10等分

```python

kf = KFold(n_splits=10, random_state=None, shuffle=False)
kf.get_n_splits(X)

print(kf)
```

### 將訓練資料隨機切割成10等分

```python
kf = KFold(n_splits=10, random_state=42, shuffle=True)
kf.get_n_splits(X)

print(kf)
```

---

### 取出 切割資料對應位置

```python
train_split = kf.split(X)
next(train_split)
```

### Or

```python
for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
```

### 取出切割資料：trainset / testset 特徵(x_train/x_test)/標註(y_train/y_test)

```python
for train_index, test_index in kf.split(X):
    X_train = X[train_index]
    Y_train = Y[train_index]
    X_test = X[test_index]
    Y_test = Y[test_index]
    print(f" \n\n X_train : {X_train} \n\n Y_train : {Y_train} \n\n X_test : {X_test} \n\n Y_test : {Y_test} \n\n")
    print("==========================================================================================================")
```
